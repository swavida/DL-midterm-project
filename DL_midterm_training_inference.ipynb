## Competition dataset

```python
# download and load competition dataset

from datasets import load_dataset

dataset = load_dataset("ad6398/nyu-dl-teach-maths-comp")

# print and see dataset

dataset
```

```python
prompt = """You are a mathematician tasked with verifying the correctness of an answer to a math question. Use the solution as context to determine whether the answer is mathematically correct. Strictly respond with 'True' if the answer is correct, otherwise 'False'.

### Question:
{}

### Solution:
{}

### Answer:
{}

### Output:
{}"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    question = examples["question"]
    solution = examples["solution"]
    ans = examples["answer"]
    output = examples["is_correct"]

    texts = []
    for instruction, step_by_step, input, output in zip(question, solution, ans, output):
        text = prompt.format(instruction, step_by_step, input, output) + EOS_TOKEN
        texts.append(text)

    return { "text" : texts }
```

```python
from datasets import load_dataset, Dataset, concatenate_datasets


total_data = dataset['train']
train_val_dataset = total_data.shuffle(seed=42).select(range(int(0.30 * len(total_data))))

# Step 2: From processed_data, split 1% for validation
split_ratio = 0.05
train_val_split_processed = train_val_dataset .train_test_split(test_size=split_ratio, seed=42)

train_dataset_full = train_val_split_processed['train']
validation_dataset = train_val_split_processed['test']

# Step 3: Balance training data to ensure equal True and False
correct_samples = train_dataset_full.filter(lambda x: x['is_correct'] == 1)
incorrect_samples = train_dataset_full.filter(lambda x: x['is_correct'] == 0)

# Determine the smaller subset size for balanced sampling
target_size = min(len(correct_samples), len(incorrect_samples))

# Sample equal number of data points from each class
balanced_correct_samples = correct_samples.shuffle(seed=42).select(range(target_size))
balanced_incorrect_samples = incorrect_samples.shuffle(seed=42).select(range(target_size))

# Combine and shuffle
balanced_train_dataset = concatenate_datasets([balanced_correct_samples, balanced_incorrect_samples]).shuffle(seed=42)


train_dataset = balanced_train_dataset.map(formatting_prompts_func, batched=True)
validation_dataset = validation_dataset.map(formatting_prompts_func, batched=True)

# Final Datasets
train_dataset, validation_dataset
```

## SFT

```python
from trl import SFTTrainer

from transformers import TrainingArguments

from unsloth import is_bfloat16_supported
from transformers import EarlyStoppingCallback


training_args = TrainingArguments(

        per_device_train_batch_size = 4,
        per_device_eval_batch_size=4,

        gradient_accumulation_steps = 4,

        warmup_steps = 200,

       #num_train_epochs = 1, # Set this for 1 full training run.

        max_steps = 4000 ,

        learning_rate = 3e-4,

        fp16 = not is_bfloat16_supported(),

        bf16 = is_bfloat16_supported(),

        logging_steps = 10,

        optim = "adamw_8bit",

        weight_decay = 0.01,

        lr_scheduler_type = "linear",

        seed = 3407,

        output_dir = "outputs",

        report_to = "wandb", # Use this for WandB etc

     save_strategy = "steps",

     save_steps = 500,
     eval_steps=500,                 # Regular validation checks to prevent overfitting

     eval_strategy="steps",    # Periodic evaluation for early stopping
     # Early Stopping
  load_best_model_at_end=True,    # Load best model at end based on validation
    metric_for_best_model="eval_loss", # Use validation loss to determine best model
  greater_is_better=False         # Lower validation loss indicates better model
    )
early_stopping_callback = EarlyStoppingCallback(early_stopping_patience=10)


trainer = SFTTrainer(

    model = model,

    tokenizer = tokenizer,

    train_dataset = train_dataset,

    dataset_text_field = "text",

    max_seq_length = max_seq_length,

    dataset_num_proc = 4,

    packing = False, # Can make training 5x faster for short sequences.
    eval_dataset=validation_dataset,
    callbacks=[early_stopping_callback],

    args = training_args

)
```

```python
test_dataset = dataset['test']

# Define prompt template
prompt_template = """You are a mathematician tasked with verifying the correctness of an answer to a math question. Use the solution as context to determine whether the answer is mathematically correct. Strictly respond with 'True' if the answer is correct, otherwise 'False'.

Now, evaluate the following:
### Question:
{}

### Solution:
{}

### Answer:
{}

### Output:
"""
prompts = [
        prompt_template.format(item['question'], item['solution'], item['answer'])
        for item in test_dataset
    ]
```

```python
prompts[0]
```

```python
FastLanguageModel.for_inference(model)

test_dataset = dataset['test']

import torch
from transformers import AutoTokenizer, BitsAndBytesConfig
import pandas as pd
import gc
from tqdm import tqdm

torch.cuda.empty_cache()
gc.collect()

# Set responses to standardize model output
true_responses = {'true', 'yes', 'correct'}
false_responses = {'false', 'no', 'incorrect'}


# Run Batch Inference
batch_size = 10  # Adjust based on available GPU memory
submission_entries = []

for i in tqdm(range(0, len(prompts), batch_size), desc="Processing batches"):
    batch_prompts = prompts[i:i + batch_size]
    inputs = tokenizer(batch_prompts, return_tensors="pt", padding=True, truncation=True).to("cuda")

    # Generate predictions
    input_token_len = inputs['input_ids'].shape[1]
    outputs = model.generate(**inputs, max_new_tokens=10, use_cache=True)

    # Decode predictions
    batch_predictions = tokenizer.batch_decode(outputs[:, input_token_len:], skip_special_tokens=True)

    # Process predictions
    for j, prediction_raw in enumerate(batch_predictions):
        normalized = prediction_raw.lower().rstrip('.').split()[0]
        predicted_label = 'True' if normalized in true_responses else 'False'

        # Save submission entry
        submission_entries.append({
            "ID": i + j,
            "is_correct": predicted_label
        })
    # Clear GPU cache after each batch
    torch.cuda.empty_cache()
    gc.collect()

# Create and save the submission CSV
submission_df = pd.DataFrame(submission_entries)
submission_df.to_csv("submission.csv", index=False)
print("Submission file 'submission.csv' created.")
```
